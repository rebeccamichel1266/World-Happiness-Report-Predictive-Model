```python
# YOUR CODE HERE
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
```

<b>Task:</b> Use the rest of this notebook to carry out your project plan. 

You will:

1. Prepare your data for your model.
2. Fit your model to the training data and evaluate your model.
3. Improve your model's performance by performing model selection and/or feature selection techniques to find best model for your problem.

Add code cells below and populate the notebook with commentary, code, analyses, results, and figures as you see fit. 

```python
# YOUR CODE HERE
features = ['Log GDP per capita', 'Social support', 'Healthy life expectancy at birth', 'Freedom to make life choices', 'Generosity', 'Perceptions of corruption', 'Positive affect', 'Confidence in national government', 'Democratic Quality', 'Delivery Quality']

target = 'Life Ladder'
```

```python
df_model = df[features + [target]].dropna()
```

```python
X = df_model[features]
y = df_model[target]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
lr = LinearRegression()
lr.fit(X_train, y_train)
y_pred_lr = lr.predict(X_test)
```

```python
print("linear Regression")
print("R2 Score:", r2_score(y_test, y_pred_lr))
print("Mean Squared Error:", mean_squared_error(y_test, y_pred_lr))
```

```python
lr_coef = pd.DataFrame({'Feature': X.columns, 'Coefficient': lr.coef_}).sort_values(by='Coefficient', ascending=False)
print("Linear Regression Coefficients:", lr_coef)
```

```python
rf = RandomForestRegressor(random_state=42)
rf.fit(X_train, y_train)
y_pred_rf= rf.predict(X_test)
```

```python
print("Random Forest")
print("R2 Score:", r2_score(y_test, y_pred_rf))
print("Mean Squared Error:", mean_squared_error(y_test, y_pred_rf))
```

```python
rf_importance = pd.DataFrame({'Feature': X.columns, 'Importance': rf.feature_importances_}).sort_values(by='Importance', ascending=False)
print("Random Forest Feature Importances:", rf_importance)
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x='Importance', y='Feature', data= rf_importance)
plt.title("Random Forest Feature Importances")
plt.show()
```
